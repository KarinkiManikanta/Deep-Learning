```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pandas_profiling
```

```python
data=pd.read_csv("/home/hemanth/Downloads/framingham_heartdisease .csv")
```

```python
data['education']=data['education'].fillna(data['education'].mean())
```

```python
data['cigsPerDay']=data['cigsPerDay'].fillna(data['cigsPerDay'].mean())
data['BPMeds']=data['BPMeds'].fillna(data['BPMeds'].mean())
data['totChol']=data['totChol'].fillna(data['totChol'].mean())
data['BMI']=data['BMI'].fillna(data['BMI'].mean())
data['heartRate']=data['heartRate'].fillna(data['heartRate'].mean())
data['glucose']=data['glucose'].fillna(data['glucose'].mean())
data.isna().sum()
```

```python
sns.pairplot(data,hue='heartRate')
```

```python
data.shape
```

```python
train_dataset=data.sample(frac=0.8,random_state=0)
test_dataset=data.drop(train_dataset.index)
```

```python
print(train_dataset.shape)
print(test_dataset.shape)
```

```python
train_stats=train_dataset.describe()
train_stats.pop('heartRate')
train_stats=train_stats.transpose()
train_stats
```

```python
train_labels=train_dataset.pop('heartRate')
test_labels=test_dataset.pop('heartRate')
```

```python
print(train_labels.shape)
print(test_labels.shape)
```

```python
def norm(x):
    return (x-train_stats['mean'])/train_stats['std']
normed_train_data=norm(train_dataset)
normed_test_data=norm(test_dataset)
```

```python
len(train_dataset.keys())
```

```python
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras import activations
from tensorflow import keras
import tensorflow as tf
```

```python
import tensorflow_docs as tfdocs
import tensorflow_docs.plots
import tensorflow_docs.modeling
```

```python
def build_model():
    model=keras.Sequential([
        layers.Dense(128,activation='relu',input_shape=[len(train_dataset.keys())]),## input layer
        layers.Dense(128,activation='relu'),## hidden layer
        layers.Dense(64,activation='relu'),
        layers.Dense(1)#output
        
      ])
    optimizer=tf.keras.optimizers.RMSprop(0.001)
    
    model.compile(loss='mse',
                 optimizer=optimizer,metrics=['mae','mse'])
    return model
```

```python
model=build_model()
model
```

```python
model.summary()
```

```python
example_batch=normed_train_data[:10]
example_result=model.predict(example_batch)
example_result
```

```python
EPOCHS = 5000

history = model.fit(
  normed_train_data, train_labels,
  epochs=EPOCHS, validation_split = 0.2, verbose=0,
  callbacks=[tfdocs.modeling.EpochDots()])
```

```python
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist.head()
```

```python
hist.tail()
```

```python
plotter = tfdocs.plots.HistoryPlotter(smoothing_std=2)
plotter
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,9))
plotter.plot({'Basic': history}, metric = "mae")
plt.ylim([0, 10])
plt.ylabel('MAE [heartRate]')
```

```python
plt.figure(figsize=(16,9))
plotter.plot({'Basic': history}, metric = "mse")
plt.ylim([0, 20])
plt.ylabel('MSE [heartRate^2]')
```

```python
model = build_model()

# The patience parameter is the amount of epochs to check for improvement
early_stop = keras.callbacks.EarlyStopping(monitor='val_loss', patience=10)

early_history = model.fit(normed_train_data, train_labels, 
                    epochs=EPOCHS, validation_split = 0.2, verbose=0, 
                    callbacks=[early_stop, tfdocs.modeling.EpochDots()])
```

```python
plotter.plot({'Early Stopping': early_history}, metric = "mae")
plt.ylim([0, 10])
plt.ylabel('MAE [heartRate]')
```

```python
loss, mae, mse = model.evaluate(normed_test_data, test_labels, verbose=2)

print("Testing set Mean Abs Error: {:10.2f} heartRate".format(mae))
```

```python
### Model Prediction
test_predictions = model.predict(normed_test_data).flatten()
plt.figure(figsize=(16,9))
a = plt.axes(aspect='equal')
plt.scatter(test_labels, test_predictions)
plt.xlabel('True Values [heartRate]')
plt.ylabel('Predictions [heartRate]')
lims = [0, 50]
plt.xlim(lims)
plt.ylim(lims)
_ = plt.plot(lims, lims)
```

```python
error = test_predictions - test_labels
plt.hist(error, bins = 100)
plt.xlabel("Prediction Error [heartRate]")
_ = plt.ylabel("Count")
```
